**(3) Exploratory Data Analysis and Visualization:**




```r
library(repr)
library(tidyverse)



player_data <- read_csv ("https://raw.githubusercontent.com/ryanceed/project-planning/refs/heads/main/players.csv")

session_data <- read_csv ("https://raw.githubusercontent.com/ryanceed/project-planning/refs/heads/main/sessions.csv")

```

```r
tidy_player_data <- player_data |>
select (experience, played_hours, Age, subscribe) |>
drop_na (Age, experience, played_hours) |>
filter(Age != 0, experience != 0, played_hours != 0) |>
mutate (experience = as_factor(experience))
head (tidy_player_data)
```

```r

player_age_plot <- tidy_player_data |>
ggplot(aes(x = Age, y = played_hours)) +
geom_point(aes(color=experience)) +
labs (x = "Age of Players (Years)", y = "Time Played (hrs)", title = "Affect of Experience and Age on Playtime", color = "Experience of Player")
player_age_plot
```

```r
player_plot2 <- tidy_player_data |>
ggplot(aes(x = subscribe, y = played_hours)) +
geom_bar(stat = "identity") +
labs (x = "Subscription status", y = "Time Played (hrs)", title = "Subscription status vs Play time")
player_plot2 
```

```r
player_plot3 <- tidy_player_data|>
ggplot(aes(x = experience, y = played_hours)) +
geom_bar(stat = "identity") +
labs (x = "Experience", y = "Time Played (hrs)", title = "Experience vs Play time") +
 theme(text = element_text(size = 12))  
  
player_plot3
```
